Parse Music Data

```python
# Import CSV File
df = pd.read_csv("features_3_sec.csv")

# Seperate columns into features
x1 = df["chroma_stft_mean"].to_numpy()
x2 = df["chroma_stft_var"].to_numpy()
x3 = df["rms_mean"].to_numpy()
x4 = df["rms_var"].to_numpy()
x5 = df["spectral_centroid_mean"].to_numpy()
x6 = df["spectral_centroid_var"].to_numpy()
x7 = df["spectral_bandwidth_mean"].to_numpy()
x8 = df["spectral_bandwidth_var"].to_numpy()

# Convert label from string to int
map = {}
y = []
for label in df['label']:
  if label not in map:
    map.update({label: len(map)})
  y.append(map.get(label))


print(y)
```

Normalize Music Data

```python
features = [x1, x2, x3, x4, x5, x6, x7, x8]

# normalize between -1 and 1
def normalize_range(arr):
  arr_min = arr.min()
  arr_max = arr.max()
  for i in range(len(arr)):
    arr[i] = (2 * (arr[i] - arr_min) / (arr_max - arr_min)) -1

for x in features:
  normalize_range(x)
```
